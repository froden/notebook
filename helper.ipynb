```kotlin
import java.awt.image.BufferedImage
import javax.imageio.ImageIO

fun convertToBlackAndWhite(inputPath: String, width: Int): BufferedImage {
    val original = ImageIO.read(File(inputPath))
    val aspect = original.width / original.height
    val height = width/aspect
    val blackAndWhite = BufferedImage(width, height, BufferedImage.TYPE_USHORT_GRAY)
    val graphics = blackAndWhite.graphics
    graphics.drawImage(original, 0, 0, width, height, null)
    graphics.dispose()
    return blackAndWhite
}
convertToBlackAndWhite("thumbs-up.jpg", 800)
```

```kotlin
val priser = DataFrame.readJson("priser.json")
priser
```

```kotlin
import kotlinx.datetime.TimeZone
import kotlinx.datetime.toLocalDateTime

val fikset = priser.convert { startsAt }.with { Instant.parse(it).toLocalDateTime(TimeZone.currentSystemDefault()).hour }
fikset
```

```kotlin
fikset.plot {
    x(startsAt)
    y(total)
    bars {
        fillColor(level) {
            scale = continuous(range = Color.YELLOW..Color.RED)
        }
    }
    line {
        y.constant(total.values.average())
        width = 2.0
        type = LineType.DASHED
    }
    layout {
        size = 1024 to 512
        theme {
            legend.position = LegendPosition.Top
        }
    }
}
```
